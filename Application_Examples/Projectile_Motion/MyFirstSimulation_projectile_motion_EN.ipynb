In this brief notebook (designed for the JupyterLab Anaconda environment), we will discuss how to calculate and plot the motion of a projectile across different launching angles. This could be your first experience using Python to simulate a physical scenario.


The problem we're addressing is illustrated in the picture below. We want to describe the motion of an object launched from a cannon at a certain angle $Î¸$.
First of all we need to choose a model for describing our projectile motion. As an assumption for our model, we neglet air friction, we suppose that the object is at the beginning at a height $h$ and has an initial velocity $v_0$ (the subscript '$0$' means here and in the following 'at $t=0$').
To describe the motion of the object, we consider separately the vertical (y) and horizontal (x) components of the motion and the case of a two dimensional motion with constant acceleration. We write the kinematics equations using the symbols $v_x$ and $v_y$ for the  velocity x- and y- components and $a_x$ and $a_y$ for their acceleration components:

For the **x-component**: \
$v_x= v_{x0} + a_x  t$ \
$x= x_{0} + v_{x0} t +\frac{1}{2}  a_x  t^2$ \
Since $a{_x}$ is equal to zero these equations becomes \
$v_x= v_{x0}$ \
$x= x_{0} + v_{x0} t$

For the **y-component**: \
$v_y= v_{y0} + a_y  t$ \
$y= y_{0} + v_{y0} t +\frac{1}{2}  a_y  t^2 = y_{0} + v_{y0} t + \frac{1}{2}  g  t^2$

where g is the gravity acceleration. \
In the code below we consider the motion for both x and y coordinates. You should now be able to follow our code based on our model.

```python
#Example: Motion of a projectile (no viscous friction of air)

import matplotlib.pyplot as plt
import math

# initial conditions
g = 9.8 #in m/s
h = 10.  # initial height in m
v0 = 5. # initial velocity in m/s
t=0. # time zero
dt=0.01  # time step

theta = 45  # angle in degree
while True:
    theta = float(input("Insert the angle theta in [0,90] degrees: "))  # With the input function the user can insert the desired angle
    if(theta>0 and theta<90): break
theta = math.radians(theta)      # convert degrees in radians

#compute velocity components
v0x = v0*math.cos(theta)
v0y = v0*math.sin(theta)
print("The initial velocity components are: \n v0_x: {:.3f} m/s \t v0_y: {:.3f} m/s".format(v0x,v0y))
# here is the reference for format method: https://www.geeksforgeeks.org/python-output-formatting/

# create lists for x and y coordinates
x=[]
y=[]
xi=0 # we initialize here the values x
yi=h  # we initialize here the values y

while yi>=0:  # we put this condition thinking about the physics of the problem
    x.append(xi)   # build the list of x position
    y.append(yi)   # build the list of y position
    t+=dt
    xi=v0x*t
    yi=h+v0y*t-0.5*g*t*t

#here we plot the trajectory of the projectile -i.e. the (x,y) motion
plt.figure(figsize=(15,7))
plt.plot(x,y, 'r.')
plt.xlabel('Range')
plt.ylabel('Height')
```

You can now choose different angles and look how the trajectory changes, or change the initial height and so on....
