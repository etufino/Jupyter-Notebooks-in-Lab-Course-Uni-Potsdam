## Graph for the first N months

We want now to do a plot of only a certain period of time. For this we use the **slicing** technique, we learnt in the second lesson about Jupyter Notebook amd Python (see Notebook).

```python
N=1000  # first N month
month,sunN = t[:N],y[:N]  # with the slicing method we choose the first N points
plt.plot(month,sunN, 'r')
plt.title("Number of Sunspots in the first {:d} months,  since January 1749 (month 0)".format(N),fontsize=22)
plt.xlabel('Months', fontsize=18)
plt.ylabel('Sunspots number',fontsize=18)
plt.show()
```

As you can see in the graph, the obtained curve is noisy. One can 'smooth' the curve by using (carefully) the method of **moving average**. This is described in the following section.

### Definition of the moving average

Suppose that $y_k$ are the measurements (in our case the sunspot numbers). We can define the moving average this way (there are also other possible definitions):  

$$Y_k = \frac{1}{2r}\sum_{m=-r}^ry_{k+m}$$

where r is an integer, the higher is its value the greater **the smoothing** of the data.
In our example below we use r=5.

#### An example of implementation in Python of the moving average defined above is the following:

```python
N=1000
r = 5
s = [ sum(y[k-r:k+r]) for k in range(N)] # we calculate the sum in the equation of the moving average
Y_moving = np.array(s)/(2*r) # we claculate the moving average and create an array of the moving average data points
print(N)
```

Now we plot the moving average data points:

```python
month,sunN = t[:N],y[:N] # selecting the data from 0 to N
plt.plot(month,sunN,'r-', label='first {:d} months'.format(N)) # we label the data using the general formatting. This way we can change N and do not have to change the legeng each time
plt.plot(month,Y_moving,'k--',label='moving average with r={:d}'.format(r),linewidth=3) # we label the data using the general formatting. This way we can change r and do not have to change the legeng each time
plt.title("Number of sunspots in the first {:d} months: original and averaged curve".format(N), fontsize=22)
plt.legend(fontsize=15)
plt.xlabel('Months', fontsize=18)
plt.ylabel('Sunspots number',fontsize=18)
plt.show()
```
