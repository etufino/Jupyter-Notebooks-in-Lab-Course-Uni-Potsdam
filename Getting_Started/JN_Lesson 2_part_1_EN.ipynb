### Slicing
An important feature when using arrays is the **slicing** of arrays. The feature allows multiple elements of an array to be called. The most basic command for the 1D array is: `[start:end:step-length]` \.
Note that the **start** is inclusive and the **end** is exclusive. If the step length is not specified, it is assumed to be 1.
If you want to output the first 4 elements of an array, write `a[0:4:1]` or `a[:4]` for short.

## Arange and linspace
Two important functions of Numpy are **arange** and **linspace**. \
**arange** generates an array of numbers between a and b with step size dx. The syntax is: `np.arange(x, b, dx)` \
**linspace** generates an array of n equidistant numbers between a and b. The syntax is: `np.linspace(a, b, n)`.

As we know, the sample standard deviation $\sigma$ can be calculated using the following formula for the variance:

$$σ_{sample}^2 = \frac{1}{N-1}\sum_{i=1}^n(x_i-\overline{x})^2$$

Here $x_i$ are the individual data points, $\overline{x}$ is the mean of the data series, and $N$ is the number of data points.

**Attention**: The function **std** with Numpy actually divides by N-ddoF (= N minus doF, i.e. N - Degree of Freedom). If you do not specify anything, the function 'np.std' defaults to ddoF=0. If we want to have the term N-1 at the denominator, we must specify ddoF=1 in the argument of the function.

## How to improve the plot (of the 1st lesson)
We can improve our plot from the last lesson by adding a legend, a title, markers and changing the font size and colors.

We will look at the same data as in the last lesson:

````python
x = [1.1, 1.5,2.4, 3.3,4.6]
y = [0.2, 0.27,0.37, 0.6,0.8]
z=[0.3,0.34,0.56,0.77,0.9]
````

````python
plt.figure(figsize=(15, 7), dpi=80)  # with 'figsize' one can change the dimensions of the figure
plt.plot(x,y,'r^--', label="data set 1")  # In 'r^', 'r' means red - '^' refers to data point symbol triangle, "--" refers to the style of line connecting the data points
plt.plot(x,z,'bo-.',linewidth=3,label="data set 2")  # In 'bo', 'b' means blue- 'o' refers to data point symbol o, '-.' refers to the dahes line connecting the data points, linewidth allows to change the line width
plt.xlabel(' Length (cm)',fontsize=16)  # x axis label , 'fontsize' command defines the font size of the x axis legend
plt.ylabel(' Force (N)',fontsize=16)   # y axis label, , 'fontsize' command defines the font size of the y axis legend
plt.title('My second plot of two data point sets', fontsize=20)  # fontsize command
plt.legend(fontsize=14) # display a legend in the plot
plt.show()
````

Plots can be dot plots without lines by using the **plt.scatter( x, y)** function instead of plt.plot(x,y).

**Attention:** Ususally, it is makes more sense to display measurement data without a connecting line!!! An exception is the representation of trajectories.
We connected the data points in this graph, just to show you how to insert a connecting line. Do not insert a connecting line when you plot your data.

### Plotting data with associated uncertainties with plt.errorbar

We can use the **plt.errorbar** function to plot data with an associated uncertainty. Below we added an uncertainty of 0.2 cm to the y-data. plt.errorbar has many options to customize the error bar display.

````python
plt.figure(figsize=(15,7))
error=0.2   # constant error in displacement
#error  = np.random.randn(len(t))*0.6   # we could generate random numbers (gaussian distributed) to simulate the uncertainties
displacement_with_error=np.array([0.4,0.7,0.8,1.1,1.4,1.9,2.3,2.7])+error
mass=[100.0,150.0,199.1,249.1,299.1,349.1,399.8,448.9]
plt.errorbar(mass,displacement_with_error,yerr=error,color='green',fmt='o',label='data')
plt.xlabel(' Mass(g)',fontsize=16)
plt.ylabel(r'$\Delta$s(cm)',fontsize=16)
plt.legend(loc='lower right', fontsize=14)
plt.grid()
plt.title("Hooke's law with error bar", fontsize=22)
plt.show()
````
