**Example:** 

Calculate the magnitude of the gravitational interaction between the earth and
you, assuming you have a mass of 60.0 kg. Consider that the earth has mass of
$5.97×10^{24}$ kg and radius 6370 km.

````python
G=6.67e-11 # gravitational constant in N*m^2/kg^2
m=60 #in kg
M_E=5.97e24
r=6370*1000.  # in meters 
F=G*m*M_E/r**2
print("Gravitational interaction in Newton=",F)
````

We will work later on the significant figures of a result.

**Exercise 2:**

Calculate the force of gravitation between the Earth and the Sun, given that the mass of the Earth is $5.97×10^{24} kg$ and that of the Sun is $1.99×10^{30} kg $.  The average distance between the two objects is $1.5×10^{11} m$ . The gravitational G constant is: $6.674×10^{−11} N*m^2⋅kg^{−2}.$

````python
#Solution
m_S=1.99e30 #mass of the sun in kg
m_E=5.97e24 #mass of the earth in kg
r=1.5e11 #distance between sun and earth in m
G=6.67e-11 #gravitational constant in N* m^2*kg^-2
F=G*(m_S*m_E)/r**2
print("Gravitational interaction in Newton=",F)
````

## Math and Numpy modules

There are several modules in Python, that allow us to use preexisting functions. You choose which module to use depending on what you need to do. The **math module** includes the most common math functions.  For reference see here: [math module](https://docs.python.org/3/library/math.html).
There is also the **Sympy module**, a module for symbolic mathematics that allows to simplify expressions, doing differentiation, matrix multiplication, etc. For more detail see here: 
[Sympy](https://docs.sympy.org/latest/index.html).

In this course, we will use math only briefly and will not use Sympy. 

In our physics laboratory course we will instead concentrate on other two very powerful modules, the so called [**NumPy**](https://numpy.org/) module (Numpy stands for Numerical Python) and the so called **SciPy** module. We will learn about SciPy later on in this course.

To use a function defined in a specific module you have to import the module and then use the following syntax:

````python
import math   # here we import the math module
math.sqrt(9)      # here we use the square root function in the math module
````

**Comment:**

The above method is the best way to import functions.
Another method would be the following:
```python
from numpy import * 
sqrt(2)
```
where the asterisk stays for 'everything'
or 
```python
from numpy import sqrt
sqrt(2)
```
This could appear a  faster method but could lead to troubles when you are using different modules in the same notebooks as you cannot be sure if you are calling sqrt function of **math** or **numpy** module.

Considering that, we strongly reccomend you to call the function in a module by using the syntax modulename.functionname.

**Example:**
