### **For** Command
Before discussing with some examples the **for** loop, we introduce the function **range** that will be very useful.

#### The function **Range**

A useful built-in function is **range**. This function returns an object that produces a sequence of integers from start point (inclusive) to end point (exclusive) by the assigned step. If not specified the default start point is zero and the default step is one. In its simplest implementation it has only one argument: the ending point.


````python
x=range(10) # the range() function returns a sequence of numbers, starting from 0 by default,
# and increments by 1 (by default), and stops before a specified number -in this case 10-.
for i in x:
  print(i)
````

**Example**:

Define a function that converts from polar coordinates (r, $\theta$) to Cartesian coordinates.
We will use sin and cos functions in the numpy package (we assume that you already imported numpy as np).


````python
def cartesian(r,theta):
  """function that converts from polar coordinates to cartesian coordinates
  theta is in rad
  """
  x= r * np.cos(theta)
  y= r * np.sin(theta)
  return (x,y)
````

````python
cartesian(1,1)  # Radius r=1 and angle $\theta$=1 rad
````

Define a function that calculate the distance between point A=$(x_A,y_A)$ and point B=$(x_B,y_B)$ . \
HInt: Use the function sqrt in the module numpy (we assume that you already imported numpy as np)


````python
#solution
def distance(pointA, pointB):
  """ function for calculate the euclidean distance
   between two points in the plane
   """
  d=np.sqrt((pointB[0]-pointA[0])**2+(pointB[1]-pointA[1])**2) # here we calculate the distance
  return round(d,3)  # round is a built in function that return a number in a given precision
distance((1,2),(3,5))  # example of using the function 'distance' to calculate the distance betweeen two points
````

**Exercise 5:**

Define and calculate the standard deviation of the four numbers you did choose above.
$$
 \sigma = \sqrt{\frac{1}{N-1} \sum_{i=1}^N (x_i - \overline{x})^2}
$$
where $\overline{x}$ is the mean of the N data.

**Check always** the numerical results!


````python
#solution
def standard_dev(x,y,z,w):
  """ It applies the standard deviation formula to calculate it for 4 numbers given as input"""
  mu=(x+y+z+w)/4  # mu is the mean
  variance=((x-mu)**2+(y-mu)**2+(z-mu)**2+(w-mu)**2)/3
  return  np.sqrt(variance)
````

````python
standard_dev(4,8,2.3,9)
````

**Example:**

Define a function  to sum the first n natural number using a while loop
As you probably know, the great scientist Gauss solved this problem (for n=100) at a very early age, by using a mathematical trick.

````python
def sum_of_n(n):
  """sum of the first n numbers, n has to be positive
  """
  sum, i=0,1    # assign initial values to the sum and to i
  while i<=n:
    sum+=i   # compact version of sum=sum+i assignment
    i +=1   ## compact version of i=i+1 assignment
  print("the sum is =", sum)
````

````python
sum_of_n(100)   # call and use the function sum_of_n
````

**Exercise 6:**

Define a function which returns the sum of the squares of the first n natural numbers.



````python
#solution
def sumsquare(n):
    """ sum of the square number, where n is positive
    """
    sum = 0
    i = 1
    while i <= n:
        sum += i**2 # compact version of sum=sum+i^2 assignment
        i += 1    # update counter

    print('The sum of the squares is', sum)

sumsquare(5)
````

## String formatting
Till now we have used the function **round()** to round a number to a given precision in decimatl digits. Instead of using the runding function, one can learn how to format the Python output of data (i.e. learn how to use the string method "format").
Suppose that the output of your exercise is a velocity of 25.43217896545 m/s stored in the variable velocity and
you want to format the output according to significant digit. With the format method you can do the following:

```
# print(" The velocity is = {:6.3f} m/s".format(velocity))
```
The output in this case will be (Try it!) :

```
# The velocity is = 25.432 m/s
```
{:6.3f} is a placeholder that controls where and in which format the value of the variable is inserted in the output string.
The format 6.3f after the colon : , indicates that the value has to be printed with 6 digits (including the decimal points) and 3 digits after the decimal points.
If you want to print an integer number, for example 514, you can use 'd' in {:d}:
```
# print("The number is:{:d}".format(velocity))

```
For printing the velocity value above in scientific notation {:6.3e}


```
# print(" The velocity is = {:6.3e} m/s".format(velocity))
```

If you are interested you can (**optionally**) review the following deeper introduction to the *format* method. The **format** method is very useful, but a little complicated and too time consuming for this course.
If you are interested in learning this, we reccomend you to consult for example the page [format method](https://python-course.eu/python-tutorial/formatted-output.php) (from where the following figure is taken) or see here for full details: https://pyformat.info

![format method](https://python-course.eu/images/python-tutorial/format_method_positional_parameters_600w.webp)


````python
# We report here the above solution with the "string" format method just explained

def volume(radius, height):
    """ radius in meter, height in meter the function return volume of cylinder """
    vol = np.pi * radius**2 * height
    return print("The volume of cylinder is={:5.2f} m^3".format(vol))  #

volume(2, 5)
````
