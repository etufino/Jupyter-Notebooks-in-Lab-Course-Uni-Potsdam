### Slicing
Eine wichtige Funktion in Verbindung mit Arrays ist das **Slicing**. Es ermöglicht durch nur einen Befehl mehrere Elemente eines Arrays aufzurufen. Die Syntax für das Slicing lautet wie folgt: `[start:ende:schrittlänge]` \
Zu beachten ist dabei, dass der Startpunkt inklusiv und der Endpunkt exklusiv ist. Wird die Schrittweite nicht anders definiert, nutzt Python eine Standard-Schrittweite von eins. Es ist außerdem möglich, nur einen Start- oder Endpunkt anzugeben. In diesem Fall würden alle Elemente ab bzw. bis zu dem festgelegten Wert ausgegeben werden.
Um die ersten vier Elemente eines Arrays zu erhalten, kann der Befehl `a[0:4:1]` oder kurz `a[:4]` genutzt werden.

### Arange und linspace
Zwei wichtige Funktionen in Numpy sind **arange** und **linspace**. \
**arange** generiert ein Array mit Zahlen zwischen einem Anfangswert a (inklusiv) und einem Endwert b (exklusive) mit der Schrittweite dx. Die Syntax lautet: `np.arange(x, b, dx)` \
**linspace** erstellt ein Array mit n äquidistanten Zahlen zwischen a (inklusiv) und b (inklusiv). Die Syntax lautet: `np.linspace(a, b, n)`

###  Mittelwert (mean) und Standardabweichung (std) in Numpy

Der Mittelwert und die Stanardabweichung einer Messreihe können problemlos mit den Numpy-Funktionen **mean** und **std** berechnet werden.
```python

```
und
```python

```

Die Standardabweichung der Stichprobe $\sigma$ kann bekanntlich mit Hilfe der folgenden Formel berechnet werden:

$$σ_{sample}^2 = \frac{1}{N-1}\sum_{i=1}^n(x_i-\overline{x})^2$$

Hierbei sind $x_i$ die Werte der individuellen Datenpunkte, $\overline{x}$ der Mittelwert der Datenreihe und $N$ die Anzahl der Datenpunkte.

Versuchen Sie, anstelle die Formel zu nutzen, den Mittelwert und die Standardabweichung direkt mit den Numpy-Funktionen **mean** und **std** zu berechnen. Geben Sie das Ergebnis im Format {:.3f} aus.

\
**Achtung**: Die Numpy-Funktion **std** dividiert durch `N-ddoF`. Dabei ist `N` die Anzahl der Elemente und `ddof` die Anzahl der Freiheitsgrade ("Delta Degrees of Freedom"). Wird nichts anderes angegeben, setzt Python `ddoF` standardmäßig auf null. Ist allerdings bspw. der Term `N-1` im Nenner gewünscht, so muss im Argument der Funktion `ddoF=1` spezifiziert werden.

### Eigenschaften von Plots
In der letzten Lektion wurde bereits ein einfacher Plot erstellt. Dieser soll nun weiter angepasst werden. Dafür können Eigenschaften wie eine Legende und ein Titel hinzugefügt oder die Markersymbole, die Schriftgröße und die Farbe geändert werden.

Die folgenden Daten entsprechen den Messreihen, welche auch im letzten Notebook genutzt wurden.

````python
x = [1.1, 1.5,2.4, 3.3,4.6]
y = [0.2, 0.27,0.37, 0.6,0.8]
z=[0.3,0.34,0.56,0.77,0.9]
````

````python
plt.figure(figsize=(15, 7), dpi=80)  # mit 'figsize' kann man die Groeße der Abbildung aendern
plt.plot(x,y,'r^--', label="Datensatz 1")
# 'r' steht fuer rot, '^' bezieht sich auf das Symbol zur Darstellung der Datenpunkte (hier ein Dreieck) und "--" auf die Art der Linie zwischen den Datenpunkten
plt.plot(x,z,'bo-.',linewidth=3,label="Datensatz 2") # 'b' steht fuer blau, 'o' bezieht sich auf das Symbol zur Darstellung der Datenpunkte (hier ein Kreis) und '-.' auf die Art der Linie zwischen den Datenpunkten, linewidth erlaubt die Aenderung der Linienbreite
plt.xlabel(' Länge (cm)',fontsize=16)  # Beschriftung der x-Achse, 'fontsize' definiert die Schriftgroeße der Legende der x-Achse
plt.ylabel(' Kraft (N)',fontsize=16)   # Beschriftung der y-Achse, 'fontsize' definiert die Schriftgroeße der Legende der y-Achse
plt.title('Mein zweiter Plot mit zwei Datenreihen', fontsize=20) # 'fontsize' definiert die Schriftgroeße des Titels
plt.legend(fontsize=14) # Anzeige einer Legende im Diagramm
plt.show() # Ausgabe des Diagramms
````

Die Messwerte lassen sich auch als Punktdiagramm ohne Verbindungslinien zwischen den Datenpunkten darstellen. Dafür wird die Funktion `plt.plot(x,y)` durch `plt.scatter( x, y)` ersetzt.


### **ACHTUNG:** In der Regel ist es sinnvoller, Messdaten ohne Verbindungslinie abzubilden. Eine Ausnahme bildet die Darstellung von Trajektorien.

### Darstellen von Daten und zugehörigen Unsicherheiten mit plt.errorbar

Mit der Funktion **plt.errorbar** können Daten inklusiv ihrer Unsicherheiten dargestellt werden.

Im Folgenden wurde eine Unsicherheit von 0,2 cm zu den y-Daten hinzugefügt. **plt.errorbar** ermöglicht viele Optionen, die Darstellung der Fehlerleiste anzupassen.

````python
plt.figure(figsize=(15,7))
Messunsicherheit=0.2   # Unsicherheit mit konstatem Wert für alle Datenpunkte
#Messunsicherheit  = np.random.randn(len(t))*0.6   # erzeugt gaussverteilte Zufallszahlen, um eine Messunsicherheiten zu simulieren
auslenkung_mit_Messunsicherheit=np.array([0.4,0.7,0.8,1.1,1.4,1.9,2.3,2.7])+Messunsicherheit
masse=[100.0,150.0,199.1,249.1,299.1,349.1,399.8,448.9]
plt.errorbar(masse,auslenkung_mit_Messunsicherheit,yerr=Messunsicherheit,color='green',fmt='o',label='data')
plt.xlabel(' Masse (g)',fontsize=16)
plt.ylabel(r'$\Delta$s(cm)',fontsize=16)
plt.legend(loc='lower right', fontsize=14)
plt.grid()
plt.title("Hooksches Gesetz mit Messunsicherheitsbalken", fontsize=22)
plt.show()
````
